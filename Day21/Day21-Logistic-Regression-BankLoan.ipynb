```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
data = pd.read_excel("Day-21/dataset/Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
```

```python
data = data.drop("ZIP Code",axis=1)
```

```python
data = data.drop("ID",axis=1)
```

```python
data
```

```python
X = data[['Age','Experience','Income','Family','CCAvg','Education','Mortgage','Securities Account','CD Account','Online','CreditCard']]
```

```python
X
```

```python
Y = data['Personal Loan']
```

```python
Y
```

```python
import statsmodels.api as sm
B0 = sm.add_constant(X)
```

```python
B0
```

```python
Logistic = sm.Logit(Y,B0)
```

```python
result = Logistic.fit()
```

```python
result.summary()
```
